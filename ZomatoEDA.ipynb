```python
df_country = pd.read_excel('Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

PIE CHART

```python
### top 3 countries that uses zomato
plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
```

Zomato maximum records and transactions are from India. After that USA and then United Kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating count'})
```

```python
ratings
```

### Observation

1. When rating is between from 4.5 to 4.9 - the rating is excellent.
2. When rating is between from 4.0 to 4.4 - the rating is very good.
3. When rating is between from 3.5 to 3.9 - the rating is good.
4. When rating is between from 2.5 to 3.4 - the rating is average.
5. When rating is between from 0.0 to 2.4 - the rating is poor.




```python
ratings.head()
```

```python
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings, hue='Rating color', palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

### Observation

1. Not rated count is very high.
2. Maximum number of rating is between 2.5 to 3.4

```python
## Count plot
sns.countplot(x='Rating color', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```
